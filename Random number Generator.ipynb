```python
import matplotlib.pyplot as plt
```

```python
def rng(m, a, c):
    rng.current = (a * rng.current + c) % m
    return rng.current / m

x=[]
# setting the seed
rng.current = 1

n_iter = 50
for i in range(n_iter):
    x.append(rng(2**16,110351,123))
plt.plot(x,marker='o')
```

# CALCULATING PI


```python
from math import sqrt
N=1000
count=0
a=[]
b=[]
for i in range(N):
    a.append(rng(2**32,23143232,2455684))
for i in range(N):  
    b.append(rng(2**23,12323429,7934634))
for i in range(N):
    r=sqrt(a[i]**2+b[i]**2)
    
    if r<1:
        count+=1
print( "{:.2f}".format(4*count/N))
```
